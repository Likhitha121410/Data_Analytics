```python
import pandas as pd
import numpy as np
order_time=[11,10,12,9]
delivery_time=[12,12,13,11]
delivery_partner=['A','B','C','D']
customer_rating=[4,5,3,2]
df=pd.DataFrame({'order_time':order_time,'delivery_time':delivery_time,'delivery_partner':delivery_partner,'customer_rating':customer_rating})
print("-------------zomato orders------------")
print(df)
print("       ")

# avg

average_delivery_time = df['delivery_time'].mean()
print(f"The average delivery time is: {average_delivery_time}")
print("       ")

#max (peak orders hours)

peak_order_hour = df['order_time'].max()
print(f"The peak order time is: {peak_order_hour}")
print("       ")

# most delay area (min)

most_delayed_area = df['delivery_time'].min()
print(f"The most delayed area is: {most_delayed_area}")
print("       ")

#top rate delivary partners

top_rated_delivery_partner = df.loc[df['customer_rating'].idxmax(), 'delivery_partner']
print(f"The top rated delivery partner is: {top_rated_delivery_partner}")
print("       ")

# visualizations

import matplotlib.pyplot as plt
plt.hist(df['delivery_time'])
plt.xlabel('Delivery Time')
plt.ylabel('Frequency')
plt.title('Distribution of Delivery Time')
plt.show()
print("       ")

plt.scatter(df['order_time'], df['delivery_time'])
plt.xlabel('Order Time')
plt.ylabel('Delivery Time')
plt.title('Order Time vs. Delivery Time')
plt.show()
print("       ")

plt.pie(df['delivery_partner'].value_counts(), labels=df['delivery_partner'].unique(), autopct='%1.1f%%')
plt.title('delivery_partner Distribution')
plt.show()
print("       ")
```
