```python

import pandas as pd
data = {
    "Patient_Name": ["likki", "shalini", "nandu", "junnu"],
    "Age": [65, 40, 72, 30],
    "Condition": ["Critical", "Stable", "Serious", "Stable"],
    "Oxygen_Level": [78, 97, 88, 95]
}

df = pd.DataFrame(data)

print("Patient Dataset")
print(df)
priority_list = []
bed_list = []
prescription_list = []

available_beds = 2

for i in range(len(df)):

    condition = df.loc[i, "Condition"]
    oxygen = df.loc[i, "Oxygen_Level"]

    # Priority
    if oxygen < 80 or condition == "Critical":
        priority = "High"

    elif oxygen < 90:
        priority = "Medium"

    else:
        priority = "Low"

    priority_list.append(priority)

    print("\nPatient:", df.loc[i, "Patient_Name"])
    print("Priority:", priority)

    # Bed Allocation
    if i < available_beds:
        bed = "Allocated"
    else:
        bed = "No Bed"

    bed_list.append(bed)

    print("Bed Allocation:", bed)

    # Prescription
    if condition == "Critical":
        prescription = "ICU Care"

    elif condition == "Serious":
        prescription = "Medication"

    else:
        prescription = "Rest"

    prescription_list.append(prescription)

    print("Prescription:", prescription)

# Add columns
df["Priority"] = priority_list
df["Bed_Allocation"] = bed_list
df["Prescription"] = prescription_list

print("\nFinal Dataset")
print(df)

import matplotlib.pyplot as plt
plt.bar(df["Patient_Name"], df["Oxygen_Level"])
plt.xlabel("Patient Name")
plt.ylabel("Oxygen Level")
plt.title("Oxygen Levels of Patients")
plt.show()

plt.scatter(df["Patient_Name"], df["Age"])
plt.xlabel("Patient Name")
plt.ylabel("Age")
plt.title("Age of Patients")
plt.show()
```
