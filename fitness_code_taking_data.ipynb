```python
from numpy import median
import pandas as pd
import statistics
import matplotlib.pyplot as plt

data=pd.read_excel("/content/gym data.xlsx")
data.drop_duplicates(inplace=True)
data.dropna(inplace=True)
data.reset_index(drop=True, inplace=True)
print(data)
print("-------------------")
average_workout=sum(data['Session_Duration (hours)'])/len(data['Session_Duration (hours)'])
print("Average Session Duration =",average_workout)
average_calories=sum(data['Calories_Burned'])/len(data['Calories_Burned'])
print("Average Calories Burned =",average_calories)
highest_calories=max(data['Calories_Burned'])
print("Highest Calories Burned =",highest_calories)
lowest_calories=min(data['Calories_Burned'])
print("Lowest Calories Burned =",lowest_calories)
median_calories=median(data['Calories_Burned'])
print("Median Calories Burned =",median_calories)
stddev=statistics.stdev(data['Calories_Burned'])
print("Standard Deviation =",stddev)
index=data['Calories_Burned'].idxmax()
print("Most Active Member has burned",data['Calories_Burned'][index],"calories")
a=plt.hist(data['Calories_Burned'])
plt.title("Calories Burned")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.show()
print("\n")
b=plt.scatter(data['Session_Duration (hours)'],data['Calories_Burned'])
plt.title("Session Duration vs Calories Burned")
plt.xlabel("Session Duration (hours)")
plt.ylabel("Calories Burned")
plt.show()
print("\n")
c=plt.boxplot(data['Calories_Burned'])
plt.title("Calories Burned")
plt.xlabel("Calories")
plt.show()
print("\n")
d=plt.bar(data['Gender'],data['Session_Duration (hours)'])
plt.title("Session Duration")
plt.xlabel("gender")
plt.ylabel("Session Duration (hours)")
plt.show()

```
