```python
import pandas as pd
df = pd.read_csv("/content/temperatures.csv")
print(df.head())
month = "JAN"
avg_temp = df[month].mean()
predicted_temp = avg_temp
print(f"\nAverage {month} Temperature: {avg_temp:.2f} °C")
print(f"Predicted {month} Temperature: {predicted_temp:.2f} °C")

import matplotlib.pyplot as plt
plt.bar(df.index, df[month])
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.title(f"Temperature in {month}")
plt.show()
print("       ")

plt.scatter(df.index, df[month])
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.title(f"Temperature in {month}")
plt.show()
```
