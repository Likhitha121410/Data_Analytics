```python
import pandas as pd
import numpy as pd

water_con=[100,50,70,95,85]
days=[4,2,1,2,3]
Tank_Cap=300
print("Tank Capacity:",Tank_Cap)

for i in range(len(water_con)):
  print(water_con[i],days[i])
  avg=sum(water_con)/len(water_con)
print("Avg: ",avg)

for i in range(len(water_con)):
  use=water_con[i]/days[i]
  print("No.of days:",days[i],"use=",use)
pred=avg+30

print("Future Pedicted Water Consumption",pred)
rem_future=Tank_Cap-pred
print("Remaining Future Water Consumption",rem_future)

import matplotlib.pyplot as plt
#bar chart
plt.bar(water_con,days)
plt.xlabel('water Consumption')
plt.ylabel('days')
plt.title('Smart water tank')
plt.show()
print("            ")
# scatter plot
plt.scatter(water_con,days)
plt.xlabel('water consumption')
plt.ylabel('days')
plt.title('Smart water tank')
plt.show()
print("     ")

# pie chart
plt.pie(
    (water_con),
    labels=(water_con)
)
plt.legend()
plt.title('water consumption')
plt.show()
```
